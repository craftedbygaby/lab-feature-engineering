There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here
spaceship = spaceship.dropna(subset=['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Name'])
```

```python
# check if dropping worked
spaceship.isnull().sum()
```

```python
spaceship.shape
```

**Perform Train Test Split**

```python
#your code here
features = spaceship.drop(columns = ['Transported'])
target = spaceship['Transported']
```

```python
# separating the test and train sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 17)
```

```python
# checking if sizes are equal
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.isnull().sum()
```

**Do transformations in Train split**

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
# perform the cleaning on train (do a function)
X_train['Cabin'].value_counts()
```

```python
# splitting column and taking only the first item and calling it for train split
def cabin_clean(df):
    df['Cabin'] = df['Cabin'].str.split('/').str[0]
    return df

X_train = cabin_clean(X_train)
```

```python
# checking if it worked
X_train['Cabin'].value_counts()
```

- Drop PassengerId and Name

```python
# creating the function to drop PII columns and calling it for train split
def anonymize(df):
    df = df.drop(columns=['PassengerId', 'Name'])
    return df
    
X_train = anonymize(X_train)
```

```python
# checking if it worked
X_train.head()
```

```python
# defining function to transform CryoSleep and VIP to boolean and calling it
def col_to_bool(df):
    df['CryoSleep'] = df['CryoSleep'].astype(bool)
    df['VIP'] = df['VIP'].astype(bool)
    return df

X_train = col_to_bool(X_train)
```

```python
# checking if it worked
X_train.dtypes
```

- For non-numerical columns, do OneHotEncoding.

```python
# non-numerical = HomePlanet, Cabin and Destination 
X_train['HomePlanet'].value_counts()
```

```python
X_train['Cabin'].value_counts()
```

```python
X_train['Destination'].value_counts()
```

```python
X_train.isnull().sum()
```

```python
# definign functions for OHE the columns, merging the new dfs into the old one and dropping the old columns
# calling the function for X-train

def OHE(df):
    df = df.reset_index(drop=True)
    encoder = OneHotEncoder(sparse_output = False)
    encoder.fit(df[['HomePlanet']])
    home_enc = encoder.transform(df[['HomePlanet']])
    df_home = pd.DataFrame(home_enc, columns = encoder.get_feature_names_out(['HomePlanet']))

    encoder.fit(df[['Cabin']])
    home_enc = encoder.transform(df[['Cabin']])
    df_cabin = pd.DataFrame(home_enc, columns = encoder.get_feature_names_out(['Cabin']))

    encoder.fit(df[['Destination']])
    home_enc = encoder.transform(df[['Destination']])
    df_dest = pd.DataFrame(home_enc, columns = encoder.get_feature_names_out(['Destination']))

    df = pd.concat([df, df_dest, df_cabin, df_home], axis=1).reset_index(drop=True)
    df = df.drop(columns=['HomePlanet', 'Cabin', 'Destination'])
    return df

X_train = OHE(X_train)
```

```python
X_train.shape
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
# defining knn and training the model
knn = KNeighborsRegressor(n_neighbors = 10)
knn.fit(X_train, y_train)
```

**Do same transformations in the Test set**

```python
# calling all functions on the test
X_test = cabin_clean(X_test)
X_test = anonymize(X_test)
X_test = col_to_bool(X_test)
X_test = OHE(X_test)
```

```python
X_test.shape
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test, y_test)
```
